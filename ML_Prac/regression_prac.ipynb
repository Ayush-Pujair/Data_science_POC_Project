# Project Admission Prediction, Lasso, Ridge Elastic Net

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
from sklearn.linear_model import Lasso, Ridge, ElasticNet, LassoCV, RidgeCV, ElasticNetCV, LinearRegression # CV here are for cross velidation
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"C:\Users\Ayush Pujari\Downloads\Admission_Prediction.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['GRE Score'] = df['GRE Score'].fillna(df['GRE Score'].mean())
```

```python
df['TOEFL Score'] = df['TOEFL Score'].fillna(df['TOEFL Score'].mean())
df['University Rating'] = df['University Rating'].fillna(df['University Rating'].mean())
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Serial No.'], inplace=True)
```

```python
# Function for all plot for the dataframe 
def plot_all(df):
    sns.boxplot(x='Research', y='Chance of Admit', data=df) # boxplot is for outliers.
    sns.swarmplot(x='Research', y='Chance of Admit', data=df) # swarmplot is for points those are there in the boxplot
    plt.show() # to show the plot
```

```python
plot_all(df)
```

```python
# Function for showing histograms for the dataframe with proper format
def plot_hist(df):
    df.hist(figsize=(20,20))
    plt.show()
```

```python
plot_hist(df)
```

```python
# Pearson's correlation graph
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap="crest", linewidths=0.4)
plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

print(arr_1.shape)

vif_score = [variance_inflation_factor(arr_1, i) for i in range(arr_1.shape[1])]
vif_column = x.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(arr_1, y, test_size=0.33, random_state=40)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
import pickle

pickle.dump(lr, open("model.pkl", "wb"))
```

```python
model = pickle.load(open("model.pkl", "rb"))
```

```python
model.get_params()
```

```python
model.intercept_
```

```python
lassocv = LassoCV(cv=20, max_iter=200000, alphas=None, random_state=42, n_jobs=-1)
lassocv.fit(X_train, y_train)
```

```python
lassocv.alpha_
```

```python
lasso = Lasso(alpha=lassocv.alpha_)
lasso.fit(X_train, y_train)
```

```python
lasso.score(X_test, y_test)
```

```python
ridgecv = RidgeCV(alphas=np.random.uniform(0,10,50), cv=50)
ridgecv.fit(X_train, y_train)
```

```python
ridgecv.alpha_
```

```python
ridgecv.score(X_test, y_test)
```

```python
ridge = Ridge(alpha=ridgecv.alpha_, random_state=22)
ridge.fit(X_train, y_train)
```

```python
ridge.score(X_test, y_test)
```

```python
elasticcv = ElasticNetCV(alphas=None, cv=50)
```

```python
elasticcv.fit(X_train, y_train)
```

```python
elasticcv.alpha_
```

```python
elasticcv.l1_ratio_
```

```python
elastic = ElasticNet(alpha=elasticcv.alpha_, l1_ratio=elasticcv.l1_ratio_)
elastic.fit(X_train, y_train)
```

```python
elastic.score(X_test, y_test)
```
